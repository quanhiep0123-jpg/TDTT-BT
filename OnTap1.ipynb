```python
def is_triangular_number(x: int) -> bool:
    if x < 1:
        return False
    
    total = 0
    n = 1
    
    while total < x:
        total += n
        if total == x:
            return True
        n += 1
    
    return False
```

```python
def factorial(n):
    gt = 1
    if(n >= 0):
        for i in range(1, n + 1):
            gt *= i
        return gt
    
    return "None"
    
assert factorial(0) == 1
assert factorial(3) == 6
assert factorial(5) == 120
assert factorial(-10) == None

print("Pass all test")
```

Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý

phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn.

```python
def fibonacci(n: int) -> int:
    if n < 0:
        return -1
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    a, b = 0, 1
    for _ in range(2, n = 1):
        a, b = b, a + b
    return b
```

### 4
 Yêu cầu: Cho các hệ số thực a, b, c, trả về tất cả các nghiệm thực của phương trình bậc hai axx + bx + c = 0 trong một danh
sách được sắp xếp tăng dần.
Quy tắc:
- Nếu a == 0 và b != 0: trả về nghiệm duy nhất [-c / b].
- Nếu a == 0 và b == 0: trả về [] (không có nghiệm hữu hạn duy nhất nếu c == 0, hoặc vô nghiệm nếu c != 0).
- Với a != 0, dùng biệt thức d = bb - 4ac:
- d < 0 → []
- d == 0 → [ -b / (2a) ]
- d > 0 → hai nghiệm [r1, r2] với r1 <= r2 

```python
import math 
def solve_quadratic(a: float, b: float, c: float) -> list[float]:
    if a == 0:
        if b == 0:
            return[]
        return[-c/b]
    delta = b*b - 4*a*c
    if delta < 0:
        return []
    elif delta == 0:
        x = -b / (2*a)
        return [x]
    else:
        sqrt_delta = math.sqrt(delta)
        x1 = (-b - sqrt_delta) / (2*a)
        x2 = (-b + sqrt_delta) / (2*a)
        return sorted([x1, x2])
a, b, c = map(int, input(). split())
print(solve_quadratic(a, b, c))
```

Định nghĩa này vẫn đúng khi có phần tử trùng nhau (ví dụ: 2, 2, 5 → giá trị giữa là 2).

```python
def middle_value(a, b, c):
    if (a <= b <= c) or (c <= b <= a):
        return b
    elif (b <= a <= c) or (c <= a <= b):
        return a
    else:
        return c
a, b, c = map(int, input(). split())
print(middle_value(a, b, c))
```

 Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3). 


```python
import math
def to_do(x1, y1, x2, y2, x3, y3):
    AB = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
    AC = ((x3 - x2)**2 + (y3 - y1)**2)**0.5
    ABxy = (x2 - x1, y2 - y1)
    ACxy = (x3 - x1, y3 - y1)
    Cos_AB_AC = ((x2 - x1)*(x3 - x1) + (y2 - y1)*(y3 - y1)) / (AB * AC)
    Cos_AB_AC = max(-1, min(1, Cos_AB_AC))
    angle_rad = math.acos(Cos_AB_AC)
    angle_deg = math.degrees(angle_rad)
    return angle_deg
x1, y1, x2, y2, x3, y3 = map(int, input().split())
print(to_do(x1, y1, x2, y2, x3, y3))
```

Yêu cầu: Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình ax^2 + bx + c = 0.
- Đầu vào: các số thực a, b, c.
- Đầu ra: 0, 1, hoặc 2. Nếu phương trình suy biến thành 0 = 0 (tức a == 0, b == 0, c == 0), trả về float('inf').
- Quy tắc:
- Nếu a != 0, dùng biệt thức D = b^2 - 4ac.
- Nếu a == 0 và b != 0, đó là phương trình bậc nhất với đúng một nghiệm thực.
- Nếu a == 0 và b == 0 và c != 0, phương trình vô nghiệm. 


```python
def quadratic_roots_count(a,b,c):
    if a==0:
        if b==0 and c==0:
            return float('inf')
        elif b==0 and c!=0:
            return ("vo_nghiem")
        elif b!=0:
            x=-c/b
            return x
    if a!=0:
        delta=b**2-4*a*c
        if delta <0:
            return ("vo_nghiem")
        elif delta==0:
            x=-b/(2*a)
            return (f"phuong trinh co nghiem kep:{x}")
        elif delta>0:
            x1=-b+(delta**0.5)/(2*a)
            x2=-b-(delta**0.5)/(2*a)
            return (f"phuong trinh co hai nghiem x1,x2=[{x1},{x2}]")

a,b,c=map(int,input().split())
print(quadratic_roots_count(a,b,c))
```

Yêu cầu: Viết hàm sum_halves(s) nhận chuỗi s gồm đúng 6 chữ số và trả về tổng của số nguyên 3 chữ số đầu và số nguyên
3 chữ số cuối.
- Đầu vào: chuỗi s độ dài 6, chỉ gồm các chữ số 0-9.
- Đầu ra: tổng kiểu số nguyên.
- Ví dụ: '100200' -> 100 + 200 = 300.
- Nếu đầu vào không hợp lệ (sai độ dài hoặc chứa ký tự không phải chữ số), phát sinh ValueError. 

```python
def sum_halves(s):
    if len(s)!=6 or not s.isdigit():
        raise ValueError
    elif len(s)==6 and s.isdigit():
        fir=int(s[:3])
        sec=int(s[3:])
        return fir+sec
    
s=input()
print(sum_halves(s))
```
