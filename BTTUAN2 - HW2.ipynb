### Bài 3:
Tính tam giác: Nhâp vào ba số a,b,c bất kì.
Kiểm tra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường.
Tính diện tích, chu vi của tam giác và in ra mànhình

```python
a, b, c = map(int, input().split())
if a+b>c and a+c>b and b+c>a:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or a == c or b == c:
        print("Tam giac can")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")
else:
    print("Khong phai tam giac")
p = (a+b+c)/2
print((p*(p-a)*(p-b)*(p-c))**0.5)
print(a+b+c)
```

### Bài 8:
Giải hệ phương trình tuyến tính:
ax + by = m
cx + dy = n

```python
a, b, c, d = map(int, input().split())
m, n = map(int, input().split())
D = a*d - b*c
Dx = m*d - b*n
Dy = a*n - m*c
if D == 0:
    if Dx == 0 and Dy == 0:
        print("Vo so nghiem")
    else:
        print("Vo nghiem")
else:
    print("Ngiem 1: ", Dx/D)
    print("Nghiem 2: ", Dy/D)
```

### Bài 9:
Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây

```python
x = int(input())
def time(x):
    if x<60:
        return str(x) + " giay"
    elif x<3600:
        return str(x//60) + " phut " + str(x%60) + " giay"
    else:
        return str(x//3600) + " gio " + str((x%3600)//60) + " phut " + str((x%3600)%60) + " giay"

print(time(x))
```
