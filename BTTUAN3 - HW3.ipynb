### Bài 14:
Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
min = 0
max = 0
while True:
    n = int(input())
    if n != -1:
        if n>max:
            max = n
        if n<min:
            min = n
        continue
    break
print(max)
print(min)
```

### Bài 19:
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó.
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input())
f1, f2 = 1, 1
while f1 + f2 <= A:
    f_next = f1 + f2
    f1 = f2
    f2 = f_next
print(f2)
```

### Bài 20:
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

```python
import math
n = int(input())
count = 0
while n != 0:
    chu_so = n % 10
    n //= 10
    if chu_so < 2:
        count += 0
    else:
        for i in range(2, math.isqrt(chu_so) + 1):
            if chu_so % i == 0:
                count += 0
        count += 1
print(count)
```
