### W4A7

Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n."

```python
n = int(input())
max_nt = 0
for i in range(2, n + 1):
    if n % i == 0:
        nt = True
        for j in range(2, i):
            if i % j == 0:
                nt = False
                break
        if nt:
            max_nt = i
print(max_nt)
```

### W4A8
ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được

```python
n = int(input())
steps = 0
def reverse_num(x):
    rev = 0
    while x > 0:
        rev = rev * 10 + x % 10
        x //= 10
    return rev
while n != reverse_num(n):
    n = n + reverse_num(n)
    steps += 1
print(steps, n)
```

```python
n = int(input())
steps = 0
while True:
    temp = n
    rev_n = 0
    while temp > 0:
        rev_n = rev_n * 10 + (temp % 10)
        temp //= 10
    if n == rev_n:
        break
    n = n + rev_n
    steps += 1
print(f"{steps} {n}")
```

### W4A9
[PerfectSquaresDistinctDigits]
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number).

```python
n = int(input())
i = 1 
found = False
while i * i <= n:
    sq = i * i
    temp = sq
    digits = []
    ok = True 
    while temp > 0:
        d = temp % 10
        if d in digits:
            ok = False 
            break
        digits.append(d)
        temp //= 10
    if ok:
        print(sq, end=" ")
        found = True
    i += 1
if not found:
    print("(no number)")
```

### W4A10
"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất."

```python
n = int(input())
max_len = 0
bets_x = 1
for x in range(1, n + 1):
    temp = x
    length = 1
    while temp != 1:
        if temp % 2 == 0:
            temp //= 2
        else:
            temp = 3 * temp + 1
        length += 1
    if length > max_len:
        max_len = length
        best_x = x
print(best_x, max_len)
```

### W4A26 [Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input())
f1, f2 = 1, 1
while f1+f2 <= A:
    f_next = f1 + f2
    f1 = f2
    f2 = f_next
print(f2)
```
