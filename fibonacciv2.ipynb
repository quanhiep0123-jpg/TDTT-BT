```python
def fib(n):
    """
    Return the n-th Fibonacci number.
    Parameters:
        n (int): The position in the Fibonacci sequence.
    Returns:
        int: The n-th Fibonacci number.
    """
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)
    # return 0
```
