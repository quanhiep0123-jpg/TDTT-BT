### W6A1
Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.

```python
numbers = list(map(int, input().split()))
result = []
for i in numbers:
    if i not in result:
        result.append(i)
print(result)
```

### W6A3
Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)
Xoay phai: Dua k ptu cuoi len dau
Xoay trai: Dua k ptu dau xuong cuoi

```python
t = tuple(map(int, input().split()))
k = int(input())
if len(t) > 0:
    k = k % len(t)
    result = t[k:] + t[:k]
else:
    result = t
print(result)
```

### W6A4
Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp lại, thêm vào list của key đó

```python
pairs = input().split()
d = {}
for pair in pairs:
    key, value = pair.split(":")
    if key not in d:
        d[key] = [value]
    else:
        d[key].append(value)
print(d)
```

### W6A5
Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)

```python
lst = list(map(int, input().split()))
d = {'positive': 0, 'negative': 0, 'zero': 0}
for i in lst:
    if i >0:
        d['positive'] += 1
    elif i < 0:
        d['negative'] += 1
    else:
        d['zero'] += 1
print(d)
```

### W6A9
Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key.

```python
d1 = eval(input())
d2 = eval(input())
d = d1.copy()
for key, value in d2.items():
    if key in d:
        d[key] += value
    else:
        d[key] = value
sorted_d = dict(sorted(d.items()))
print(sorted_d)
```

### W6A10
Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k.
In ra list gồm các cặp số khác nhau có tổng bằng k.
Mỗi cặp la một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai.
Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple

```python
lst = list(map(int, input().split()))
k = int(input())
result = []
n = len(lst)
for i in range(n):
    for j in range(i, n):
        if lst[i] + lst[j] == k:
            result.append((lst[i], lst[j]))
result.sort()
print(result)
```

### W6A11 [Tách số chẵn và lẻ từ tuple]
Nhập vào một tuple chứa các số nguyên.
Hãy tạo ra hai tuple mới: một chứa tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên thứ tự tương đối.
In ra hai tuple này trên hai dòng riêng biệt.

```python
t = tuple(map(int, input().split()))
chan = []
le = []
for i in t:
    if i % 2 == 0:
        chan.append(i)
    else:
        le.append(i)
print(tuple(chan))
print(tuple(le))
```

### W6A12 [Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list.
Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất trong số đó

```python
lst = list(map(int, input().split()))
lst.sort()
max_count = 0
for i in lst:
    count = lst.count(i)
    if count > max_count:
        max_count = count
        result = i
print(result)
```

### W6A13 [Đảo ngược key và value của dictionary]

### W6A14 [Giao của hai danh sách]
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên.
Hãy tìm các phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các phần tử chung đó, không trùng lặp.

```python
lst1 = list(map(int, input().split()))
lst2 = list(map(int, input().split()))
lst = []
for i in lst1:
    if i in lst2:
        if i not in lst:
            lst.append(i)
print(lst)
```

### W6A15 [Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số nguyên k ở dòng tiếp theo.
Hãy tạo một dictionary mới chỉ chứa các cặp key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới

```python
d = eval(input())
k = int(input())
d_new = {}
for key, value in d.items():
    if value > k:
        d_new[key] = value
print(d_new)
```
